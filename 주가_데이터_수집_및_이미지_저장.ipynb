```python
# ==========================================
# [주식 트랙] [1] 주가 데이터 수집 및 이미지 저장 (최종)
# ==========================================

!pip install finance-datareader mplfinance

import FinanceDataReader as fdr
import mplfinance as mpf
import pandas as pd
import matplotlib.pyplot as plt # 메모리 관리를 위해 추가
import os
from google.colab import drive

# 1. 구글 드라이브 연결
drive.mount('/content/drive')

# 2. 저장 경로 설정
# 경로 수정: 본인의 프로젝트 폴더명에 맞게 확인
save_path = '/content/drive/MyDrive/P실무프로젝트_3조/images'
os.makedirs(save_path, exist_ok=True)
print(f"이미지 저장 경로: {save_path}")

# 3. 데이터 불러오기
ticker = '005930' # 삼성전자
# 팁: 딥러닝 학습용이라면 기간을 2018년 정도로 늘리는 것을 추천합니다.
start_date = '2023-01-01'
end_date   = '2023-12-31'

print("데이터 불러오는 중...")

try:
    df = fdr.DataReader(ticker, start_date, end_date)

    # 인덱스가 DatetimeIndex인지 확인 (안전장치)
    if not isinstance(df.index, pd.DatetimeIndex):
        df.index = pd.to_datetime(df.index)

    print(f"{ticker} 데이터 {len(df)}개 로드 완료!")
except Exception as e:
    print("데이터 로드 실패:", e)
    raise

# 4. 차트 생성 스타일 설정
mc = mpf.make_marketcolors(up='r', down='b', inherit=True)
style = mpf.make_mpf_style(marketcolors=mc)

window_size = 20
print("차트 이미지 변환 시작!")

count = 0
N = len(df) - window_size

for i in range(N):
    subset = df.iloc[i : i + window_size]

    # 파일명: 윈도우의 마지막 날짜 기준 (예: 1월1일~1월20일 데이터 -> 1월20일.png)
    target_date = subset.index[-1].strftime('%Y-%m-%d')
    filename = f"{save_path}/{target_date}.png"

    # 이미지가 없을 때만 생성 (중복 방지)
    if not os.path.exists(filename):
        try:
            mpf.plot(
                subset,
                type='candle',
                style=style,
                axisoff=True,
                volume=False,
                savefig=dict(fname=filename, dpi=50, pad_inches=0)
            )
            # [추가] 메모리에 쌓인 그래프 찌꺼기 제거 (대량 생성 시 중요)
            plt.close('all')

            count += 1
        except Exception as e:
            print(f"[WARN] {target_date} 이미지 생성 실패:", e)

    # 진행률 표시
    if i % 50 == 0:
        print(f"{i}/{N} 진행 중... ({target_date})")

print(f"\n완료! 총 {count}장의 이미지가 저장되었습니다.")
```
